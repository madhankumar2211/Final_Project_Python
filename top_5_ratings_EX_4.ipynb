```python
import pandas as pd
import os
```

```python
ratings = pd.read_csv(os.getcwd() + "/ratings.csv", encoding='utf8')
movies = pd.read_csv(os.getcwd() + '/movies.csv', encoding='utf8')
```

```python
rating_count = ratings.groupby('movieId').count()
rating_index = rating_count[rating_count>4].dropna().index.tolist()
rating_index
```

```python
rating_df = pd.DataFrame()
rating_df['avg_rating'] = ratings.groupby('movieId').rating.mean()
```

```python
rating_df = rating_df[rating_df.index.isin(rating_index)]
```

```python
movies_ratings = movies.merge(rating_df,left_on='movieId',right_index=True,how='left').dropna()
```

```python
movies_ratings
```

```python
movies_ratings['genres'] = movies_ratings['genres'].str.split('|')
movies_ratings = movies_ratings.explode('genres')
```

```python
movies_ratings
```

```python
final_df = movies_ratings.sort_values(by = ['genres','avg_rating','title'],ascending = (True,False,True)).groupby('genres').head(5)
```

```python
final_df = final_df.round({'avg_rating':2})
```

```python
final_df.to_csv(os.getcwd()+'/top5.csv',index=False)
```
